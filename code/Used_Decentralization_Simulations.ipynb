#Fixed Lambda

```python
#Using dataframes
def fix_lambda(lm):
  df = pd.DataFrame(0, index=np.arange(10000), columns=['trans_val'])
  ent = pd.DataFrame(0, index=np.arange(10000), columns=['entropy','n'])
  for n in range(1,10001):
    v = np.exp(lm*n/10**7)
    df.loc[n] = v
    df['pr'] = df['trans_val']/df['trans_val'].sum()
    H = -df['pr']*np.log2(df['pr'])
    E = H.sum()
    ent.loc[n] = 2**E
    ent.loc[n]['n'] = n
  return ent
```

```python
lambd0 = fix_lambda(0)
lambd1 = fix_lambda(5000)
lambd2 = fix_lambda(10000)
```

```python
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=lambd0['n'], y=lambd0['entropy'],
              mode='lines', name='Lambda = 0'))
fig1.add_trace(go.Scatter(x=lambd1['n'], y=lambd1['entropy'],
              mode='lines', name='Lambda = 5000'))
fig1.add_trace(go.Scatter(x=lambd2['n'], y=lambd2['entropy'],
              mode='lines', name='Lambda = 10000'))
fig1.update_layout(title='Decentralization Index with Variable N at Fixed Lambda',
                   xaxis_title='Number of Total Transactions',yaxis_title='Index Value',font=dict(family="Courier New, monospace", size=18, color="black"),margin=go.layout.Margin(
        l=0, #left margin
        r=0, #right margin
        b=0, #bottom margin
        t=0  #top margin
    ))
fig1.show()
fig1.write_image('FixedNSimulation.pdf',scale=3)
```

#Fixed N

```python
#Using dataframes
def fix_n(n):
  df = pd.DataFrame(0, index=np.arange(n), columns=['trans_val'])
  ent = pd.DataFrame(0, index=np.arange(100000/1000), columns=['entropy','lambda'])
  i = 0
  df['trans_val'] = np.exp((df.index.values+1)/10**9)
  for lm in range (0,100000,1000):
    df['trans_tmp'] = df['trans_val']**lm
    df['pr'] = df['trans_tmp']/df['trans_tmp'].sum()
    H = -df['pr']*np.log2(df['pr'])
    E = H.sum()
    ent.loc[i] = 2**E
    ent.loc[i]['lambda'] = lm
    i += 1
  #ent.rename(columns={'entropy':'{}'.format(n)},inplace=True)
  return ent
```

```python
n100000 = fix_n(100000)
n50000 =fix_n(50000)
n10000 = fix_n(10000)
n10 = fix_n(100)
```
